```python
import pandas as pd
```

```python
csv = pd.read_csv("./Android-Apps/Google-Playstore.csv")
csv.dropna()
```

```python
csv.head()
```

```python
csv['Content Rating'].unique()
```

```python
csv = csv.drop(
    [
        'Rating',
        'Rating Count',
        'Installs',
        'Minimum Installs',
        'Maximum Installs',
        'Free',
        'Price',
        'Developer Website',
        'Developer Email',
        'Released',
        'Last Updated',
        'Privacy Policy',
        'Ad Supported',
        'In App Purchases',
        'Editors Choice',
        'Scraped Time',
        'Category',
        'Currency',
        'Size',
        'Minimum Android',
        'Developer Id',
    ],
    axis=1
)
```

```python
csv.head()
```

```python
unique_content_ratings = csv["Content Rating"].unique()
unique_content_ratings
```

```python
csv.head()
```

```python
age_group_map = {
    "Unrated": 0,
    "Everyone": 1,
    "Everyone 10+": 10,
    "Teen": 13,
    "Mature 17+": 17,
    "Adults only 18+": 18,
}

content_ratings_map = {
    "Unrated": 0,
    "Everyone": 1,
    "Everyone 10+": 2,
    "Teen": 3,
    "Mature 17+": 4,
    "Adults only 18+": 5,
}


for content_rating in unique_content_ratings:
    csv.loc[csv['Content Rating'] == content_rating, 'Age'] = age_group_map[content_rating]
    csv.loc[csv['Content Rating'] == content_rating, 'Content Rating'] = content_ratings_map[content_rating]


```

```python
csv.head()
```

```python
csv.tail()
```

```python
csv.describe()
```

```python
csv['App'] = csv['App Name']
csv['Package Name'] = csv['App Id']
csv['Age-appropriate'] = csv['Content Rating']
```

```python
csv = csv.drop(['App Name', 'App Id'], axis=1)
csv = csv[['App','Package Name','Content Rating', 'Age']]
```

```python
csv['Age'] = csv['Age'].astype('int')
```

```python
csv = csv[csv['App'].str.contains(",") == False]
```
